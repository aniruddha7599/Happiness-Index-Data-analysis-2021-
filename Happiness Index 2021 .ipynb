# Happiness Index 

## Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd 
%matplotlib inline
```

## Reading CSV file

```python
df=pd.read_csv('2021.csv')
df
```

## Sorting rows for India 

```python
India=df[df['Country name']=='India']
India
```

## Statistical data for India

```python
India.describe()
```

## Plot for "Life ladder" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Life Ladder'],linestyle='-',marker='*',color='darkorange')
plt.title("Life Ladder")
plt.xlim(2005,2021)
plt.ylim(3,5.5)
plt.grid()
plt.show()
```

## Plot for "GDP per capita" vs"year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Log GDP per capita'],linestyle='-',marker='o',color='red')
plt.title("Log GDP per capita")
plt.xlim(2005,2021)
plt.ylim(8,9,0.1)
plt.grid()
plt.show()
```

## Plot for "Social Support" vs"year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Social support'],linestyle='-.',marker='*',color='green')
plt.title("Social support")
plt.xlim(2005,2021)
plt.ylim(0.500,0.750)
plt.grid()
plt.show()
```

## Plot for "Healthy life expectancy at birth" vs"year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Healthy life expectancy at birth'],linestyle=':',marker='*',color='magenta')
plt.xlim(2005,2021)
plt.ylim(55,62)
plt.title("Healthy life expectancy at birth")
plt.grid()
plt.show()
```

## Plot for "Freedom to make life choices" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Freedom to make life choices'],linestyle='--',marker='o',color='black')
plt.title("Freedom to make life choices")
plt.xlim(2005,2021)
plt.ylim(0.55,0.95)
plt.grid()
plt.show()
```

## Plot for "Generosity" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Generosity'],linestyle='solid',marker='o',color='brown')
plt.title("Generosity")
plt.xlim(2005,2021)
plt.ylim(-0.0850,0.120)
plt.grid()
plt.show()
```

## Plot for "Perceptions of corruption" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Perceptions of corruption'],linestyle=':',marker='o',color='steelblue')
plt.title("Perceptions of corruption")
plt.xlim(2005,2021)
plt.ylim(0.72,0.92)
plt.grid()
plt.show()
```

## Plot for "Positive Effect" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Positive affect'],linestyle='-.',marker='o',color='olive')
plt.title("Positive affect")
plt.xlim(2005,2021)
plt.ylim(0.62,0.78)
plt.grid()
plt.show()
```

## Plot for "Negative Effect" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Negative affect'],linestyle='-',marker='o',color='gray')
plt.title("Negative affect")
plt.xlim(2005,2021)
plt.ylim(0.18,0.48)
plt.grid()
plt.show()
```

## Plot for "Happines Index" vs "year"

```python
plt.figure(figsize=(20,7))
plt.plot(India['year'],India['Happines Index '],marker='o',color='orange')
plt.title("Happines Index ")
plt.xlim(2005,2021)
plt.ylim(8,8.7)
plt.grid()
plt.show()
```
